```python
import numpy as np
import matplotlib.pyplot as plt

params = {"axes.labelsize": 14, "axes.titlesize": 14}
plt.rcParams["axes.linewidth"] = 1
plt.rcParams['figure.dpi'] = 100
plt.rcParams.update(params)

def place(ax):
    ax.tick_params(direction="in", which="major", length=5, labelsize=13)
    ax.tick_params(direction="in", which="minor", length=3)
    ax.grid(which="major", ls="--", dashes=(1, 3), lw=0.8, zorder=0)

data = np.loadtxt("rdf.dat", skiprows=4)

r    = data[:, 1]
g_CC = data[:, 2]
g_CO = data[:, 4]
g_OO = data[:, 6]

fig, ax = plt.subplots(figsize=(6, 4))

for g, label in zip([g_CC, g_CO, g_OO], ["C-C", "C-O", "O-O"]):
    ax.plot(r, g, lw=2, label = label)
    ax.set_xlabel(r"$r$ (Å)")
    ax.set_ylabel(r"$g(r)$")
    place(ax)

plt.legend()
plt.tight_layout()
plt.savefig("rdf.png", dpi=300, bbox_inches="tight")
plt.show()
```
